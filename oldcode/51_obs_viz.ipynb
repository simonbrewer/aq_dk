## Maps

```python
import geopandas as gpd
```

```python
aq_df = pd.read_csv("./data/aqs_data.csv")
aq_df.head()
```

```python
aq_crds = aq_df.loc[:, ['longitude', 'latitude', 'site.num']]
aq_crds = aq_crds.drop_duplicates()
```

```python
aq_gdf = gpd.GeoDataFrame(
    aq_crds, geometry=gpd.points_from_xy(aq_crds.longitude, aq_crds.latitude), crs="EPSG:4326"
)
```

```python
aq_gdf
```

```python
ax = aq_gdf.plot()

for x, y, label in zip(aq_gdf.geometry.x, aq_gdf.geometry.y, aq_gdf['site.num']):
    ax.annotate(label, xy=(x, y), xytext=(3, 3), textcoords="offset points")
```

```python
import contextily as cx
aq_gdf_wm = aq_gdf.to_crs(epsg=3857)
ax = aq_gdf_wm.plot(figsize=(10, 10))
for x, y, label in zip(aq_gdf_wm.geometry.x, aq_gdf_wm.geometry.y, aq_gdf_wm['site.num']):
    ax.annotate(label, xy=(x, y), xytext=(3, 3), textcoords="offset points")
cx.add_basemap(ax)
```
