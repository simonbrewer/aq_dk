# Basis functions and scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

import xgboost as xgb
```

## Load data

```python
dat = pd.read_csv("./rbfs/phi_short_reduce.csv")
dat.head()
```

```python
X = dat.iloc[:, 1:218].to_numpy()
y = dat['y'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = 0.8, random_state=42)
```

```python
xgb_model = xgb.XGBRegressor(objective="reg:linear", random_state=42)
```

```python
aq_xgb = xgb_model.fit(X_train, y_train)
```

```python
y_test2 = y_test.reshape(1, -1)
```

```python
y_test_pred = aq_xgb.predict(X_test)
```

```python
plt.plot(y_test, y_test_pred, 'bo')
plt.axline((0,0), slope=1)
plt.show()
```

```python
print(f"MAE: {mean_absolute_error(y_test, y_test_pred):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_test_pred):.4f}")
```
